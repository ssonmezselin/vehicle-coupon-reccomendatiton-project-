# In-vehicle Coupon Recommendation  System With ANN

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('datasetrcouponrec.csv')
```

```python
dataset.head()
```

```python
X = dataset.drop(labels=['destination', 'passanger','time','temperature','toCoupon_GEQ5min','weather','Y','car','Bar','CoffeeHouse','CarryAway','RestaurantLessThan20','Restaurant20To50'], axis = 1)
y = dataset['Y']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label1 = LabelEncoder()
X['maritalStatus'] = label1.fit_transform(X['maritalStatus'])
```

```python
X.head()
```

```python
label = LabelEncoder()
X['gender'] = label.fit_transform(X['gender'])
X.head()
```

```python
label = LabelEncoder()
X['coupon'] = label.fit_transform(X['coupon'])
X['occupation'] = label.fit_transform(X['occupation'])
X['expiration'] = label.fit_transform(X['expiration'])
X['income'] = label.fit_transform(X['income'])
X['age'] = label.fit_transform(X['age'])
X['education'] = label.fit_transform(X['education'])
X.head(20)
```

### Feature Standardization 

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 10)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

### Build ANN 

```python
model = Sequential()
model.add(Dense(X.shape[1], activation='relu', input_dim = X.shape[1]))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
X.shape[1]
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train.to_numpy(), batch_size = 32, epochs = 1000, verbose = 1)
```

```python
y_pred = model.predict_classes(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.evaluate(X_test, y_test.to_numpy())
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
